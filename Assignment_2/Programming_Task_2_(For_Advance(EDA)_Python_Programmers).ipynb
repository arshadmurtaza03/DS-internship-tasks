## Q1 - Defanging an IP Address

```python
class Solution:
    def defangIPaddr(self, address: str) -> str:
        return address.replace(".", "[.]")
```

## Q2 - Find Numbers with Even Number of Digits

```python
from typing import List

class Solution:
    def findNumbers(self, nums: List[int]) -> int:
        even_digit_count = 0

        for number in nums:
            digits = len(str(number))

            if digits % 2 == 0:
                even_digit_count += 1

        return even_digit_count
```

## Q3 - Number of Good Pairs

```python
class Solution:
    def numIdenticalPairs(self, nums: List[int]) -> int:
        pair_count = 0
        seen = {}

        for num in nums:
            if num in seen:
                pair_count += seen[num]
                seen[num] += 1
            else:
                seen[num] = 1

        return pair_count
```

## Q4 - How Many Numbers Are Smaller Than the Current Number

```python
class Solution:
    def smallerNumbersThanCurrent(self, nums: List[int]) -> List[int]:
        result = []

        for i in range(len(nums)):
            count = 0

            for j in range(len(nums)):
                if nums[j] < nums[i]:
                    count += 1

            result.append(count)

        return result

```

## Q5 - Subtract the Product and Sum of Digits of an Integer

```python
class Solution:
    def subtractProductAndSum(self, n: int) -> int:
        digit_sum = 0
        digit_product = 1

        while n > 0:
            digit = n % 10
            digit_sum += digit
            digit_product *= digit
            n //= 10

        return digit_product - digit_sum
```

## Q6 - Subtract the Product and Sum of Digits of an Integer

```python
class Solution:
    def xorOperation(self, n: int, start: int) -> int:
        result = 0

        for i in range(n):
            value = start + 2 * i
            result ^= value

        return result

```
