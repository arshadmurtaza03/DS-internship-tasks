# Q6- Wite a function

```python
def is_leap(year):
    leap = False

    # Write your logic here

    if year % 4 != 0:
        leap = leap

    else:
        if year % 100 !=0:
            leap = not leap

        else:
            if year % 400 !=0:
                leap = leap

            else:
                leap = not leap

    return leap

year = int(input())
print(is_leap(year))
```
